```python

import os
import nltk
from nltk.corpus import wordnet as wn
from nltk import pos_tag, word_tokenize
from collections import Counter


nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

def load_texts(folder_path):
    texts = []
    
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()
                texts.append(content)
                
                print(f"Loaded: {filename}")
    
    print(f"Total files: {len(texts)}\n")
    return texts
    def process_texts(texts):
    tagged_words = []
    
    for text in texts:
        tokens = word_tokenize(text)
        tagged = pos_tag(tokens)
        tagged_words.extend(tagged)
    
    print(f"Total words processed: {len(tagged_words)}")
    return tagged_words

    def get_most_common(words, n=10):
    counter = Counter(words)
    common = counter.most_common(n)
    
    for word, count in common:
        print(f"{word}: {count}")
    
    return common

    def explore_wordnet(word):
    synsets = wn.synsets(word)
    data = []
    
    for syn in synsets:
        data.append({
            "definition": syn.definition(),
            "examples": syn.examples()
        })
    
    return data


def analyze_words(words, limit=10):
    results = {}
    
    unique_words = list(set(words))[:limit]
    
    for word in unique_words:
        results[word] = explore_wordnet(word)
    
    return results

    def save_results(filename, results):
    with open(filename, 'w', encoding='utf-8') as f:
        
        for word, entries in results.items():
            f.write(f"WORD: {word}\n")
            
            for entry in entries:
                f.write(f"  Definition: {entry['definition']}\n")
                
                for example in entry['examples']:
                    f.write(f"  Example: {example}\n")
            
            f.write("\n-----------------\n\n")
    
    print(f"Saved to {filename}")

   
season1_path = "transcripts/season1"
season2_path = "transcripts/season2"


season1_texts = load_texts(season1_path)
season2_texts = load_texts(season2_path)


season1_tagged = process_texts(season1_texts)
season2_tagged = process_texts(season2_texts)

season1_nouns = extract_words(season1_tagged, "NN")
season2_nouns = extract_words(season2_tagged, "NN")


print("\n--- Season 1 Most Common Nouns ---")
get_most_common(season1_nouns, 10)

print("\n--- Season 2 Most Common Nouns ---")
get_most_common(season2_nouns, 10)


season1_results = analyze_words(season1_nouns, 10)
season2_results = analyze_words(season2_nouns, 10)


save_results("season1_wordnet.txt", season1_results)
save_results("season2_wordnet.txt", season2_results)

season1_results = analyze_words(season1_nouns, 10)
season2_results = analyze_words(season2_nouns, 10)


save_results("season1_wordnet.txt", season1_results)
save_results("season2_wordnet.txt", season2_results)
```
